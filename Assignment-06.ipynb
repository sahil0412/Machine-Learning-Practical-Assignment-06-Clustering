```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import numpy as np
```

```python
df = pd.read_csv("data_stocks.csv")
```

```python
df.head()
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], unit='s')
```

```python
df.head()
```

```python
stocks_data = df.iloc[:,2:]
```

```python
stocks_data.head()
```

```python
returns = stocks_data.pct_change()
returns.head()
```

```python
ann_ret = returns.mean()*252*100
print("STOCK Annual Return %\n")
print(ann_ret)
```

```python
ann_vol = returns.std()*np.sqrt(252)*100
print("STOCK Annual Volatility %\n")
print(ann_vol)
```

```python
annual = pd.DataFrame()
annual["Return"] = returns.mean()*252
annual["Volatility"] = returns.std() * np.sqrt(252)
annual
```

## Optimal Clusters using Elbow Curve

```python
from sklearn.cluster import KMeans

data = np.asarray([np.asarray(annual['Return']),np.asarray(annual['Volatility'])]).T

X = data
distorsions = []
for k in range(2, 20):
    k_means = KMeans(n_clusters=k)
    k_means.fit(X)
    distorsions.append(k_means.inertia_)
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

## K-Means Clustering with K = 6

```python
from pylab import plot,show
from scipy.cluster.vq import kmeans,vq

# computing K-Means with K = 6 (6 clusters)
centroids,_ = kmeans(data,6)
# assign each sample to a cluster
idx,_ = vq(data,centroids)
# some plotting using numpy's logical indexing
plot(data[idx==0,0],data[idx==0,1],'ob',
     data[idx==1,0],data[idx==1,1],'oy',
     data[idx==2,0],data[idx==2,1],'or',
     data[idx==3,0],data[idx==3,1],'og',
     data[idx==4,0],data[idx==4,1],'om',
    data[idx==5,0],data[idx==5,1],'oc')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()
```
